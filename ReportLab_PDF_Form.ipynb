```
from reportlab.lib.pagesizes import A4
from reportlab.platypus import (
    SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle, Image, PageTemplate, Frame
)
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib import colors
from reportlab.pdfbase.cidfonts import UnicodeCIDFont
from reportlab.pdfbase import pdfmetrics
from reportlab.lib.units import inch

# Register a font for Unicode support
# Make sure the font 'HeiseiMin-W3' is available to ReportLab
try:
    pdfmetrics.registerFont(UnicodeCIDFont('HeiseiMin-W3'))
    FONT_NAME = 'HeiseiMin-W3'
except Exception:
    print("Warning: 'HeiseiMin-W3' font not found. Falling back to 'Helvetica'.")
    FONT_NAME = 'Helvetica'


# Define the PDF path (changed to relative path)
pdf_path = "Mimo_Bank_Account_Details_Form.pdf"

# Create the document
doc = SimpleDocTemplate(
    pdf_path,
    pagesize=A4,
    rightMargin=40,
    leftMargin=40,
    topMargin=60,
    bottomMargin=60,
)

# Brand colors
PLANTATION = colors.HexColor("#255347")
CHINO = colors.HexColor("#CDC29F")
SUSHI = colors.HexColor("#76AF2C")
CASCADE = colors.HexColor("#92A5A4")

# Styles
styles = getSampleStyleSheet()

header_style = ParagraphStyle(
    name="Header",
    fontName=FONT_NAME,
    fontSize=18,
    leading=22,
    alignment=1,  # Centered
    textColor=PLANTATION,
    spaceAfter=20,
)

label_style = ParagraphStyle(
    name="Label",
    fontName=FONT_NAME,
    fontSize=12,
    leading=16,
    textColor=PLANTATION,
)

footer_style = ParagraphStyle(
    name="Footer",
    fontName=FONT_NAME,
    fontSize=10,
    leading=14,
    alignment=1,
    textColor=CASCADE,
)

# Build the header with a logo placeholder
# NOTE: The <img ...> tag in a Paragraph will not render an image file.
# It will just print the literal text of the tag.
# To add an image, you should use the 'Image' object (which you've imported).
# Example:
# try:
#     logo = Image("your-logo-path.png", width=70, height=70)
#     header = [logo, Spacer(1, 12), Paragraph("Bank Account...", header_style)]
# except:
#     print("Logo file not found. Using text only.")
#     header = [Spacer(1, 12), Paragraph("Bank Account...", header_style)]

header = [
    Spacer(1, 12),
    # This line will print the literal text "<img src=...>"
    Paragraph('<img src="your-logo-path.png" width="70" height="70"/>', header_style),
    Paragraph("Bank Account Details Form – Mimo’s Collective", header_style),
]

# Data fields
fields = [
    ("Full Name", "Movsum Mirzazada"),
    ("Company Name (optional)", "Mimo’s Collective"),
    ("Email Address", "mimo@movsummirzazada.com"),
    ("Phone Number", "+971585929669"),
    ("Bank Name", "Emirates NBD"),
    ("Account Holder Name", "Mirmovsum Mirzazada"),
    ("Account Number", "1015713470301"),
    ("IBAN", "AE020260001015713470301"),
    ("SWIFT/BIC Code", "EBILAEAD"),
    ("Bank Address", "Emirates NBD Building – Jumeira St – opposite Jumeira Beach Hotel – Umm Suqeim 3 – Dubai, UAE"),
    ("Currency", "AED"),
    ("Nationality", "Azerbaijani"),
    ("ID/Passport Number", "________________________"),  # Placeholder
    ("Upload ID Document", "________________________"),  # Placeholder for uploading
    ("Signature", "________________________"),  # Placeholder for digital/typed signature
]

# Create the table
table_data = [
    [
        Paragraph(f"<b>{label}:</b>", label_style),
        Paragraph(value, label_style),
    ]
    for label, value in fields
]

form_table = Table(table_data, colWidths=[2.5 * inch, 3.5 * inch])
form_table.setStyle(
    TableStyle(
        [
            # ('BACKGROUND', (0, 0), (-1, 0), CHINO), # This style was targeting the first row
            # ('TEXTCOLOR', (0, 0), (-1, 0), PLANTATION), # This style was targeting the first row
            ('GRID', (0, 0), (-1, -1), 1, CASCADE), # Added a grid for clarity
            ('VALIGN', (0, 0), (-1, -1), 'TOP'),
            ('BACKGROUND', (0, 0), (0, -1), CHINO), # Style first column background
            ('TEXTCOLOR', (0, 1), (-1, -1), colors.black), # Changed data text to black for readability
            ('LINEABOVE', (0, 0), (-1, 0), 2, PLANTATION), # Thicker line above first row
            ('LINEBELOW', (0, -1), (-1, -1), 2, PLANTATION), # Thicker line below last row
        ]
    )
)

# Footer information
footer = Paragraph(
    "Movsum Mirzazada | mimo@collective | www.movsummirzazada.bio", footer_style
)

# Elements to build
elements = header + [Spacer(1, 20), form_table, Spacer(1, 30), footer]

# Build the document
try:
    doc.build(elements)
    print(f"Successfully generated PDF: {pdf_path}")
except Exception as e:
    print(f"Error generating PDF: {e}")
```
